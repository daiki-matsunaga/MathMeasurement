```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 20.0, 0.01)

R = 1.0
r = 0.15
rho = 0.10

x = (R - r)*np.cos(t) + rho*np.cos((R - r)/r*t)
y = (R - r)*np.sin(t) + rho*np.sin((R - r)/r*t)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x[0], y[0], "ro")
ax.set_aspect("equal")
plt.show()
```
